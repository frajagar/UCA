```python
#BASIC SCRIPTS

def cons(k,l):#array constante de término k y longitud l
    x=[]
    for i in range(l):
        x=x+[k]
    return(x)

def zero(l):#array constante de término 0 y longitud l
    x=[]
    for i in range(l):
        x=x+[0]
    return(x)

def extract(A):#extrae, del array de arrays A, un array con el primer elemento de cada array de A
    E=[]
    for i in range(len(A)):
        E=E+[A[i][0]]
    return E

def addJ(A,J):#añade J a las componentes del array A
    B=[]
    for j in range(len(A)):
        B=B+[A[j]+J]
    return(B)
    
def col(A,j):#extrae un array con el elemento j de cada array de un array A
    column=[]
    for i in range(len(A)):
        if j+1<=len(A[i]):
            column=column+[A[i][j]]
        else:
            column=column+[]
    return(column)

def switch(X,j,x):#cambia la componente j del array X por x 
    Y=[]
    for i in range(len(X)):
        if i==j:
            Y=Y+[x]
        else:
            Y=Y+[X[i]]
    return(Y)

def switchcol(A,j,C):#cambia la columna j del array de arrays A por la columna C
    B=[]
    for i in range(len(A)):
        B=B+[switch(A[i],j,C[i])]
    return(B)   

def superswitch(X,J,C):#va cambiando recursivamente la columna J[j] del array de arrays A por las componente C[j] del array C
    Y=[]
    for j in range(len(J)):
        X=switch(X,J[j],C[j])
    return(Y)

def lc(W):#devuelve la longitud del array más grande del array de arrays W
    l=[]
    for n in range(len(W)):
        l=l+[len(W[n])]
    L=max(l)
    return(L)

def trans(W):#transpone el array de arrays W
    l=lc(W)
    O=[]
    for h in range(0,l):
        O=O+[col(W,h)]
    return(O)
    
def tale(X,k):#corta la cola del array X a partir del índice k
    Y=[]
    for i in range(0,k+1):
        Y=Y+[X[i]]
    return(Y)

def inter(G,H):#pone dos arrays G y H como columnas adyacentes y vuelta a array
    K=[]
    for i in range(min(len(G),len(H))):
        K=K+[[G[i],H[i]]]
    return(K)

def scaprod(X,Y):#producto escalar de dos arrays X e Y
    l=min(len(X),len(Y))
    s=0
    for i in range(l):
        s=s+X[i]*Y[i]
    return(s)

def triscaprod(X,Y,Z):#producto escalar de tres arrays X, Y y Z
    l=min(len(X),len(Y),len(Z))
    s=0
    for i in range(l):
        s=s+X[i]*Y[i]*Z[i]
    return(s)

def joint(U,V):#suma dos arrays U y V componente a componente
    T=[]
    le=min(len(U),len(V))
    for l in range(le):
        T=T+[U[l]+V[l]]
    return(T)

def algprod(X,Y):#producto, componente a componente, de los arrays X e Y
    Z=[]
    le=min(len(X),len(Y))
    for k in range(le):
        Z=Z+[X[k]*Y[k]]
    return(Z)

def fnorm(X):#suma las componentes del array X
    if X==[]:
        x=0
    else:
        x=X[0]
        for i in range(len(X)-1):
            x=x+X[i+1]
    return(x)

def SS(A,k):#suma las componentes k-ésimas de los arrays del array A
    s=0
    for i in range(len(A)):
        s=s+A[i][k]
    return(s)
    
def act(X,x):#multiplica el array X por el coeficiente x
    Y=[]
    for j in range(len(X)):
        Y=Y+[X[j]*x]
    return(Y)
    
def canonical(x,i,l):#genera el array canónico de longitud l y valor x en la posición i
    y=[]
    for j in range(l):
        if j==i:
            y=y+[x]
        else:
            y=y+[0]
    return(y)
    
def add(X,x,i):#añade un array canónico de valor x en la posición i al array X
    X=joint(X,canonical(x,i,len(X)))
    return(X)
    
def gadd(X,Y,J):#va añadiendo un array canónico de valor Y[j] en la posición J[j] al array X
    Z=X
    for j in range(len(J)):
        Z=add(Z,Y[j],J[j])
    return(Z)
    
def scanonical(x,i,l,k):#genera un array constante de valor k excepto x en la posición i
    y=[]
    for j in range(l):
        if j==i:
            y=y+[x]
        else:
            y=y+[k]
    return(y)
    
def sadd(X,x,i,k):#añade un array constante de valor k excepto x en la posición i al array X
    X=sumjoint(X,scanonical(x,i,len(X),k))
    return(X)
    
def sgadd(X,Y,J):#va añadiendo un array constante de valor X[0] excepto Y[j] en la posición J[j] al array X
    Z=X
    for j in range(len(J)):
        Z=sadd(Z,Y[j],J[j],X[0])
    return(Z)
    
def sumjoint(E,F):#suma los arrays de arrays E y F
    l=min(len(E),len(F))
    G=[]
    for i in range(l):
        G=G+[joint(E[i],F[i])]
    return(G)
    
def simpcomp(T,f):#aplica la función de una variable f a las componentes del array T
    D=[]
    for h in range(len(T)):
        D=D+[f(T[h])]
    return(D)
    
def doubcomp(T,f):#aplica recursivamente la función de dos variables f a cada par de componentes del array T
    D=T[0]
    if len(T)==1:
        D=D
    else:
        if len(T)==2:
            D=f(T[0],T[1])
        else:
            for h in range(len(T)-2):
                D=f(T[0],T[1])
                D=f(D,T[h+2])
    return(D)
    
def simpdoubcomp(C,S,f):#aplica la función de dos variables f al par de componentes i-ésimas de los arrays C y S
    l=min(len(C),len(S))
    P=[]
    for i in range(l):
        P=P+[f(C[i],S[i])]
    return(P)

def subarray(A,a,b):#extrae aquellos arrays del array A cuya componente a-ésima es b
    SUB=[]
    for l in range(len(A)):
        if A[l][a]==b:
            SUB=SUB+[A[l]]
        else:
            SUB=SUB
    return(SUB)

def dsubarray(A,T,b):#extrae aquellos arrays del array A cuya componente [T[0]][T[1]]-ésima es b
    DSUB=[]
    for l in range(len(A)):
        if A[l][T[0]][T[1]]==b:
            DSUB=DSUB+[A[l]]
        else:
            DSUB=DSUB
    return(DSUB)

def pre(A,i):#corta el array A desde la componente i en adelante
    B=[]
    for j in range(len(A)):
        if j<i:
            B=B+[A[j]]
        else:
            B=B
    return(B)

def post(A,i):#se come la primera parte del array A hasta la componente i
    B=[]
    for j in range(len(A)):
        if j>i:
            B=B+[A[j]]
        else:
            B=B
    return(B)

def retract(A,l):#se come los primeros elementos hasta el l de cada array del array de arrays A
    E=[]
    for i in range(len(A)):
        E=E+[post(A[i],l)]
    return E

def substract(A,i):#extrae la componente i del array A
    B=pre(A,i)+post(A,i)
    return(B)

def position(A,a):#array de las posiciones de la componente a en el array A
    J=[]
    for i in range(len(A)):
        if A[i]==a:
            J=J+[i]
        else:
            J=J
    return(J)

def superposition(A,B):#crea un array con las posiciones de las componentes del array B en el array A
    J=[]
    for i in range(len(B)):
        J=J+[position(A,B[i])]
    return(J)

def replace(Z,j,z):#reemplaza la componente j del array Z por z
    NZ=[]
    for k in range(len(Z)):
        if k==j:
            NZ=NZ+[z]
        else:
            NZ=NZ+[Z[k]]
    return(NZ)

def clean(Z):#elimina componentes repetidos del array Z
    V=[]
    for w in range(len(Z)):
        if len(position(Z,Z[w]))==1:
            V=V+[Z[w]]
        else:
            Z=replace(Z,w,[])
    return(joint(V))

def reindex(V,I):#permuta las componentes del array V según el array de índices I
    W=[]
    for i in range(len(I)):
        W=W+[V[I[i]]]
    return(W)

def order(X):#ordena el array X de menor a mayor
    maxX=max(X)
    RX=[]
    J=[]
    for i in range(len(X)):
        minX=min(X)
        Q=position(X,minX)
        RX=RX+[minX]
        X=substract(X,Q[0])
    return(RX)

def orderind(X):#saca la permutación de índices al ordenar el array X
    RX=order(X)
    Q=[]
    for i in range(len(X)):
        if (len(Q)<len(X)):
            P=position(X,RX[i])
            Q=Q+P
        else:
            Q=Q
    return(Q)

def ordercol(X,b):#ordena el array de arrays X por la columna b
    C=col(X,b)
    D=order(C)
    I=orderind(C)
    RX=reindex(X,I)
    OC=[RX,I]
    return(OC)
    
def superorder(X,b):#ordena el array de arrays X por la columna b
    Ba=col(col(X,6),b)
    O=order(Ba)[0]
    SP=superposition(Ba,O)
    In=clean(SP)
    RX=reindex(X,In)
    return(RX)

def tablita(rows,columnlabels,rowlabels):#hace una tabla con las filas rows con las etiquetas de columnas columnlabels y etiquetas de filas rowlabels
    data=np.array(rows)
    if rowlabels=='':
        df=pd.DataFrame(data,columns=columnlabels)
        with pd.option_context("display.max_rows",None,"display.max_columns",None):
            display(df)
    else:
        df=pd.DataFrame(data,columns=columnlabels,index=rowlabels)
        with pd.option_context("display.max_rows",None,"display.max_columns",None):
            display(df)
    return(df)

def tablitaS(rows,columnlabels,rowlabels):#hace la tablita con una última fila de sumas
    SP=[SS(rows,h) for h in range(len(columnlabels))]
    rows=rows+[SP]
    rowlabels=rowlabels+['TOTAL']
    tablita(rows,columnlabels,rowlabels)

def tablitaSS(rows,columnlabels,rowlabels):#hace la tablita con una última fila de sumas
    rows=[rows[h]+[fnorm(rows[h])] for h in range(len(rows))] 
    columnlabels=columnlabels+['TOTAL']
    tablita(rows,columnlabels,rowlabels)

def tablitaSSS(rows,columnlabels,rowlabels):#hace la tablita con una última fila de sumas
    rows=[rows[h]+[fnorm(rows[h])] for h in range(len(rows))] 
    columnlabels=columnlabels+['TOTAL']
    tablitaS(rows,columnlabels,rowlabels)

def dropdown(opt,des,dp):
    @interact 
    def _(code=widgets.Dropdown(options=['']+extract(opt),value='',description=des,disabled=False)):
        for i in range(len(opt)):
            if code==extract(opt)[i]:
                dp
    return(None)

def victor(X):#calcula las posiciones del máximo del array X
    M=max(X)
    P=position(X,max(X))
    return(P)
```

```python
#DSOA SCRIPTS

def conv(A):#convierte el array de códigos de actividades en un array numérico
    B=[]
    for i in range(len(A)):
        for j in range(len(labelACT)):
            if A[i]==labelACT[j]:
                B=B+[j]
    return(B)

def calsim(matric,danger,coe):#recalcula la simultaneidad si no hay suficientes estudiantes
    if danger=='medium':
        if ((coe==4)&(matric>=15))or((coe>=5)&(matric>=7)):
            cs=2
        else:
            cs=1
    else:
        cs=1
    return(cs)

def sizeD(coe,sim):#tamaño de grupos por coeficiente de experimentalidad y simultaneidad
    if coe==1:
        s=40
    else:
        if coe==2:
            s=35
        else:
            if coe==3:
                s=30
            else:
                if coe==4:
                    if sim>=2:
                        s=30
                    else:
                        s=20
                else:
                    if coe==5:
                        if sim>=2:
                            s=25
                        else:
                            s=15
                    else:
                        if coe==6:
                            s=10
                        else:
                            s=0
    return(s)

def sizeE(coe,sim):#tamaño de grupos por coeficiente de experimentalidad y simultaneidad
    if 1<=coe<=3:
        s=40
    else:
        if 4<=coe<=6:
            s=20
        else:
            s=0
    return(s)

def capnorm(matric,coe,sim,act):#capacidad de un grupo por normativa
    cn=[75,40,25,sizeD(coe,sim[fnorm(position(act,3)+[0])]),sizeE(coe,sim[fnorm(position(act,4)+[0])]),5,12,40,20,6,20,75]
    return(cn)

def hedasize(matric,siz):#hedas por tamaño de grupos
    He=[0.4,0.8,1.0,1.2,1.6,2.0,2.4]
    if (siz>=51)&(siz<=75):
        si=He[0]
    else:
        if (siz>=76)&(siz<=100):
            si=H[1]
        else:
            if (siz>=101)&(siz<=125):
                si=H[2]
            else:
                if (siz>=126)&(siz<=150):
                    si=H[3]
                else:
                    if (siz>=151)&(siz<=175):
                        si=H[4]
                    else:
                        if (siz>=176)&(siz<=200):
                            si=H[5]
                        else:
                            if (siz>=201):
                                si=H[6]
                            else:
                                si=0
    return(si)

def roundgrupnorm(alum,q):#redondea número de grupos por normativa según matriculados
    if alum==0:
        g=0
    else:
        if math.modf(alum)[1]==0:
            g=1
        else:
            if math.modf(alum)[0]>=q:
                g=math.modf(alum)[1]+1
            else:
                g=math.modf(alum)[1]
    return(int(g))

def roundprof(profe):#redondea número de profesores
    if math.modf(profe)[0]>=0.5:
        profe=math.modf(profe)[1]+1
    else:
        profe=math.modf(profe)[1]
    return(profe)

def sectionlabel(arel):#etiqueta las columnas del desglose de una asignatura en grupos de actividad
    L=labelsec
    for h in range(len(arel)):
        L=L+[arel[h]]
    return(L)

def asignacion(arec,ects,hp,grup,sim):#asignación total de las áreas de una asignatura
    asi=[]
    for i in range(len(arec)):
        asi=asi+[triscaprod(arec[i][1],grup,sim)*hp]
    return(asi)

def sections(arec,arel,coe,matric,ects,hp,grup,sim,act):#grupos de actividad de una asignatura
    SECT=[]
    grupnorm=[]
    HP=[]
    for i in range(len(labelact)):
        sect=[]
        grupnorm=grupnorm+[roundgrupnorm(matric/capnorm(matric,coe,sim,act)[i],0.4)]
        HP=HP+[ects[i]*hp*grup[i]*sim[i]]
        sect=sect+[capnorm(matric,coe,sim,act)[i],grupnorm[i],ects[i],grup[i],sim[i],HP[i]]
        for j in range(len(arec)):
            sect=sect+[arec[j][1][i]*hp*grup[i]*sim[i]]
        SECT=SECT+[sect]
    return(SECT+[cons('',2)+[fnorm(ects)]+cons('',2)+[fnorm(HP)]+asignacion(arec,ects,hp,grup,sim)])

def arecc(arec,act):#créditos ects por área en una asignatura
    ectsc=[]
    for i in range(len(arec)):
        ectsc=ectsc+[gadd(zero(12),arec[i][1],act)]
    ARECC=switchcol(arec,1,ectsc)
    return(ARECC)

def asignatura(ID,arec,sem,matri,act,ects,hp,grup,sim):#asignaturas de las titulaciones
    are=extract(arec)
    arel=col(are,0)
    coe=are[0][1]
    asig=asignacion(arec,ects,hp,grup,sim)
    areasig=inter(arel,asig)
    matric=fnorm(matri)
    arec=arecc(arec,act)
    ects=gadd(zero(12),ects,act)
    grup=gadd(zero(12),grup,act)
    sim=gadd(zero(12),sim,act)
    sec=sections(arec,arel,coe,matric,ects,hp,grup,sim,act)
    ASIGNATURA=[ID,areasig,matri,sec,sem,arel,fnorm(ects)]
    return(ASIGNATURA)

def framesubject(ASIGNA):#genera el dataframe de una asignatura
    rows=ASIGNA[3]
    columnlabels=sectionlabel(ASIGNA[5])
    rowlabels=labelact+['']
    return(tablita(rows,columnlabels,rowlabels))

def ndisplay(ASIG):#muestra el desglose en grupos de actividad de una array de asignaturas
    for a in range(len(ASIG)):
        print(ASIG[a][0])
        rows=ASIG[a][3]
        columnlabels=sectionlabel(ASIG[a][5])
        rowlabels=labelact+['']
        tablita(rows,columnlabels,rowlabels)
    return(None)

def ndisplaycsv(ASIG):#muestra el desglose en grupos de actividad de una array de asignaturas
    print(ASIG[0])
    rows=ASIG[3]
    columnlabels=sectionlabel(ASIG[5])
    rowlabels=labelact+['']
    return(tablita(rows,columnlabels,rowlabels))

def gradog(ID,CENTRO,AREAS,ASIGNATURAS):#array de un grado con su docencia grupal
    GR=grupa(AREAS,ASIGNATURAS)
    GRL=grupal(AREAS,ASIGNATURAS)
    ASIGNATURAS1=asigsem(ASIGNATURAS,1)
    GR1=grupa(AREAS,ASIGNATURAS1)
    GRL1=grupal(AREAS,ASIGNATURAS1)
    ASIGNATURAS2=asigsem(ASIGNATURAS,2)
    GR2=grupa(AREAS,ASIGNATURAS2)
    GRL2=grupal(AREAS,ASIGNATURAS2)
    GRADO=[ID,CENTRO,AREAS,ASIGNATURAS,GR,GRL,ASIGNATURAS1,GR1,GRL1,ASIGNATURAS2,GR2,GRL2]
    return(GRADO)

def AREC(AREAS,titu,ects):#construye el array de áreas y sus pesos en una titulación
    areC=[]
    for i in range(len(AREAS)):
        areC=areC+[[AREAS[i],[weight(AREAS[i],titu)*ects[j] for j in range(len(ects))]]]
    return(areC)

def gradogi(gradog,ASIGNATURASI):#completa el array de un grado con la docencia individual
    IN=grupa(gradog[2],ASIGNATURASI)
    INL=grupal(gradog[2],ASIGNATURASI)
    TOTA=grupat(gradog[2],gradog[3],ASIGNATURASI)
    TOTAL=grupalt(gradog[2],gradog[3],ASIGNATURASI)
    GRADO=gradog+[ASIGNATURASI,IN,INL,TOTA,TOTAL]
    return(GRADO)

def asigsem(AS,q):#asignacion en un semestre
    T=[4,1]
    ASQ=dsubarray(AS,T,q)
    return(ASQ)

def edaras(AR,AS):
    ENCDOC=0
    for k in range(len(AS)):
        for l in range(len(AS[k][1])):
            if (AS[k][1][l][0]==AR[0]):
                ENCDOC=ENCDOC+AS[k][1][l][1]
            else:
                ENCDOC=ENCDOC
    return(ENCDOC)

def grupa(AREAS,ASIG):
    GR=[]
    for z in range(len(AREAS)):
        GR=GR+[[AREAS[z],edaras(AREAS[z],ASIG)]]
    return(GR)

def grupat(AREAS,ASIG,ASIL):
    GR=[]
    for z in range(len(AREAS)):
        GR=GR+[[AREAS[z],edaras(AREAS[z],ASIG),edaras(AREAS[z],ASIL)]]
    return(GR)

def grupal(AREAS,ASIG):
    GRL=[]
    for z in range(len(AREAS)):
        GRL=GRL+[[AREAS[z][0],edaras(AREAS[z],ASIG)]]
    return(GRL)

def grupalt(AREAS,ASIG,ASIL):
    GRL=[]
    for z in range(len(AREAS)):
        GRL=GRL+[[AREAS[z][0],edaras(AREAS[z],ASIG),edaras(AREAS[z],ASIL)]]
    return(GRL)
    
def carga(Titulacion,p):#carga horaria-profesor de una titulación p=4,7,10,13
    Carga=0
    for h in range(len(Titulacion[p])):
        Carga=Carga+Titulacion[p][h][1]
    return Carga

def edt(Area,Titulacion,p):#encargo docente de un área en una titulación p=4,7,10,13
    EncargoDocente=0
    for j in range(len(Titulacion[p])):
        if Titulacion[p][j][0]==Area:
            EncargoDocente=EncargoDocente+Titulacion[p][j][1]
        else:
            EncargoDocente=EncargoDocente
    return EncargoDocente

def weight(ar,tit):#peso de un área en una titulación
    we=edt(ar,tit,4)/carga(tit,4)
    return(we)

def edc(Area,Centro,Titulaciones,p):#encargo docente de un área en un centro p=4,7,10,13
    EncargoDocente=0
    for j in range(len(Titulaciones)):
        if Titulaciones[j][1]==Centro:
            EncargoDocente=EncargoDocente+edt(Area,Titulaciones[j],p)
        else:
            EncargoDocente=EncargoDocente
    return EncargoDocente

def edca(Area,Campus,Centros,Titulaciones,p):#encargo docente de un área en un campus p=4,7,10,13
    EncargoDocenteCA=0
    for n in range(len(Centros)):
        if Centros[n][1]==Campus:
            EncargoDocenteCA=EncargoDocenteCA+edc(Area,Centros[n],Titulaciones,p)
        else:
            EncargoDocenteCA=EncargoDocenteCA
    return EncargoDocenteCA

def edtt(Area,Titulaciones,p):#encargo docente de un área en la universidad p=4,7,10,13
    EncargoDocenteT=0
    for j in range(len(Titulaciones)):
        EncargoDocenteT=EncargoDocenteT+edt(Area,Titulaciones[j],p)
    return EncargoDocenteT

def pra(ECTS):#horas profesor de un practicum
    if (ECTS>=3)&(ECTS<=5):
        h=0.1
    else:
        if (ECTS>=6)&(ECTS<=11):
            h=0.2
        else:
            if (ECTS>=12)&(ECTS<=17):
                h=0.4
            else:
                if (ECTS>=18):
                    h=0.6
                else:
                    h=0
    return(h/ECTS)

def balare(T):#calcula el balance de un área
    bal=zero(len(areas))
    for i in range(len(T)):
        for j in range(len(T[i][1])):
            for k in range(len(areas)):
                if areas[k][0]==T[i][1][j][0]:
                    bal=add(bal,T[i][1][j][1],k)
    return(bal)

def hedasex(SEX,year):#HEDAs por sexenios
    if year<=SEX[1]+6:
        if SEX[0]==1:
            hs=40
        else:
            if SEX[0]==2:
                hs=60
            else:
                if SEX[0]>=3:
                    hs=80
                else:
                    hs=0
    else:
        if SEX[0]>=5:
            hs=80
        else:
            hs=0
    return(hs)

def capfinal(CI,HEDAS):#capacidad final
    cf=CI-fnorm(HEDAS)
    cflosu=max(cf,120)
    cfuca=CI-HEDAS[3]
    if (HEDAS[3]==140)or(HEDAS[3]==160)or(HEDAS[3]==180)or(HEDAS[3]==240):
        cff=cfuca
    else:
        cff=cflosu
    return(cff)

def coestr(areas):#calcula la estructuralidad de las áreas
    Rareas=[]
    for b in range(4):
        Rareas=Rareas+[superorder(areas,b)]
    COE=[]
    for i in range(len(Rareas)):
        COE=COE+[[Rareas[i][0],Rareas[i][6][1],Rareas[i][6][2],Rareas[i][6][3],Rareas[i][6][4]]]
    return(COE)

def coestr2(areas):#muestra la estructuralidad de las áreas
    Rareas=[]
    for b in range(4):
        print(col(superorder(areas,b),0))
    return(None)

def situarea(area,PDI):#situación de un área
    EDG1=[edtt(area,grados,7),edtt(area,masteres,7),edtt(area,doctorados,7)]
    EDG2=[edtt(area,grados,10),edtt(area,masteres,10),edtt(area,doctorados,10)]
    EDG=[edtt(area,grados,4),edtt(area,masteres,4),edtt(area,doctorados,4)]
    EDI=[edtt(area,grados,13),edtt(area,masteres,13),edtt(area,doctorados,13)]
    ED=[EDG1,EDG2,EDG,EDI]
    CIE=fnorm(col(PDI[0],5))
    CINE=fnorm(col(PDI[1],5))
    CI=[CIE,CINE]
    HEDADOC=fnorm(col(col(fnorm(PDI),6),0))
    HEDASEX=fnorm(col(col(fnorm(PDI),6),1))
    HEDAINV=fnorm(col(col(fnorm(PDI),6),2))
    HEDAGES=fnorm(col(col(fnorm(PDI),6),3))
    HEDAINT=fnorm(col(col(fnorm(PDI),6),4))
    HEDAFOR=fnorm(col(col(fnorm(PDI),6),5))
    HEDAPER=fnorm(col(col(fnorm(PDI),6),6))
    HEDAS=[HEDADOC,HEDASEX,HEDAINV,HEDAGES,HEDAINT,HEDAFOR,HEDAPER]
    CFE=fnorm(col(PDI[0],7))
    CFNE=fnorm(col(PDI[1],7))
    CF1=CIE+CINE
    CF2=CIE
    CF3=CIE+CINE-HEDASEX
    CF4=CIE+CINE-HEDASEX-HEDAPER
    CF5=CIE+CINE-fnorm(HEDAS)
    CF=[CFE,CFNE,CF1,CF2,CF4,CF5]
    BAL0=CFE+CFNE-fnorm(EDG)
    BAL1=CF1-fnorm(EDG)
    BAL2=CF2-fnorm(EDG)
    BAL3=CF3-fnorm(EDG)
    BAL4=CF4-fnorm(EDG)
    BAL5=CF5-fnorm(EDG)
    BAL=[BAL0,BAL1,BAL2,BAL3,BAL4,BAL5]
    area=area+[ED,PDI,CI,HEDAS,CF,BAL]
    return(area)

def plantilla(area):#plantilla de un área
    PL=[labelplantilla]
    pl=fnorm(area[3])
    for i in range(len(pl)):
        PL=PL+[[pl[i][0],pl[i][1],pl[i][2][0],pl[i][3][0],pl[i][4][0],pl[i][4][1],pl[i][5],pl[i][6][0],pl[i][6][1],pl[i][6][2],pl[i][6][3],pl[i][6][4],pl[i][6][5],pl[i][6][6],pl[i][7],pl[i][8]]]
    show(table(PL))
    return(None)

def plantillac(area,centro):#plantilla de un área en un centro
    PL=[labelplantilla]
    pl=fnorm(area[3])
    for i in range(len(pl)):
        if pl[i][2]==centro:
            PL=PL+[[pl[i][0],pl[i][1],pl[i][2][0],pl[i][3][0],pl[i][4][0],pl[i][4][1],pl[i][5],pl[i][6][0],pl[i][6][1],pl[i][6][2],pl[i][6][3],pl[i][6][4],pl[i][6][5],pl[i][6][6],pl[i][7],pl[i][8]]]
        else:
            PL=PL
    show(table(PL))
    return(None)

def plantillaca(area,campus):#plantilla de un área en un campus
    PL=[labelplantilla]
    pl=fnorm(area[3])
    for i in range(len(pl)):
        if pl[i][2][1]==campus:
            PL=PL+[[pl[i][0],pl[i][1],pl[i][2][0],pl[i][3][0],pl[i][4][0],pl[i][4][1],pl[i][5],pl[i][6][0],pl[i][6][1],pl[i][6][2],pl[i][6][3],pl[i][6][4],pl[i][6][5],pl[i][6][6],pl[i][7],pl[i][8]]]
        else:
            PL=PL
    show(table(PL))
    return(None)

def committee(M,H,p):#devuelve un array de dos componentes que son variaciones de M y H de longitud p, respectivamente
    lm=len(M)
    lh=len(H)
    m=random.sample(range(0,lm),p)
    h=random.sample(range(0,lh),p)
    RM=[]
    for i in range(p):
        RM=RM+[M[m[i]]]
    RH=[]
    for i in range(p):
        RH=RH+[H[h[i]]]
    R=[RM,RH]
    return(R)
    
def compco(B,k,p):#devuelve un array donde cada componente tiene dos componentes de variaciones de longitud p de elementos de B
    S=[]
    for m in range(len(B)):
        S=S+[committee(B[m][k][0],B[m][k][1],p)]
    return(S)

def hedascentros(E):#dada la estructura E de un centro y las correspondientes hedas H, te calcula las hedas totales
    H=[140,90,90,90,60,60,60]
    hc=scaprod(E,H)
    return(hc)

def year(Y):#convierte años, meses y días a años
    y=Y[0]+Y[1]/12+Y[2]/365
    return(y)

def sabbatical(candidato):#calcula la puntuación de un sabático
    t=2*candidato[1]
    Pesocargo=[1,0.6,0.3,0.2,0.1]
    Years=[]
    for h in range(len(candidato[3])):
        W=fnorm(simpcomp(candidato[3][h],year))
        Years=Years+[W]
    Cargos=algprod(Pesocargo,Years)
    CAN=[candidato[0],t,candidato[2]]+Cargos
    return(CAN)

def sabbareport(Candidatos):#tablita de puntuaciones de los sabáticos
    Candidatos=simpcomp(Candidatos,sabbatical)
    rows=[col(Candidatos,j+1) for j in range(len(Candidatos[0])-1)]
    columnlabels=extract(Candidatos)
    rowlabels=labelsabba
    tablitaS(rows,columnlabels,rowlabels)
    return(None)

def rcd(nhp,nbal):#necesidad de certificado de consejo de departamento para imformes de viabilidad
    if (nhp>0)&(nbal<=-90):
        ccd='SI'
    else:
        ccd='NO'
    return(ccd)

def deficit(D):#deficit estructural
    defi=[]
    if min(D)>=0:
        defi=defi+[D[len(D)-1]]
    else:
        for i in range(len(D)):
            if D[i]<0:
                defi=defi+[D[i]]
    return(defi[0])

def newpads(nde,nhp):#nuevas PADs
    if (nde<-90)&(nhp>0):
        NewPADs=roundprof(nhp/180)
    else:
        NewPADs=0
    return(NewPADs)

def escenarios(CCD,NPADs):#escenarios de informes de viabilidad
    if position(CCD,'SI')==[]:
        print('CONCLUSIÓN: ESCENARIO A')
    else:
        print('CONCLUSIÓN: ESCENARIO B')
    if max(NPADs)>=1:
        print('CONCLUSIÓN: ESCENARIO C')
    
def feasibility(Titu,areasreport):#informe de viabilidad de áreas en una titulación
    display(Image(url='https://gabcomunicacion.uca.es/wp-content/uploads/2017/05/Logo-V1-Color-Web-600x237-pix-gif.gif?u',width=300,height=100,unconfined=True))
    print('INFORME DEL VICERRECTORADO DE PROFESORADO SOBRE VIABILIDAD DE MODIFICACIÓN DE MEMORIAS DE TITULACIONES O IMPLANTACIÓN DE NUEVAS TITULACIONES')
    print('')
    print('Curso 25-26')
    print(Titu[0])
    print(Titu[1][0])
    print('')
    rows=[]
    CCD=[]
    NPADs=[]
    for l in range(len(areasreport)):
        nhp=areasreport[l][7]
        bal=areasreport[l][6]
        nbal=bal-nhp
        ccd=rcd(nhp,nbal)
        CCD=CCD+[ccd]
        de=deficit([areasreport[l][2],areasreport[l][3],areasreport[l][4],areasreport[l][5]])
        nde=de-nhp
        npads=newpads(nde,nhp)
        NPADs=NPADs+[npads]
        rows=rows+[[nhp,bal,nbal,ccd,de,nde,npads]]
    columnlabels=labelreport
    rowlabels=extract(areasreport)
    tablita(rows,columnlabels,rowlabels)
    print('')
    print('ESCENARIOS:')
    print('A: Cada área en la que aumentan las horas-profesor tiene suficiente balance para asumirlas (nuevo balance superior o igual a -90).')
    print('B: Existen áreas donde aumentan las horas-profesor que no tienen suficiente balance  para asumirlas (nuevo balance inferior -90), por lo que es necesario certificado del Consejo de Departamento con el compromiso de asumir dichas horas sin contratación extra de profesorado.')
    print('C: Existen áreas donde aumentan las horas-profesor de manera que su estructuralidad se ve afectada hasta el punto de necesitar nuevas plazas de ayudante doctor y por tanto una nueva reordenación de la lista de estructuralidad de las áreas.')
    print('')
    print('Nota: Obsérvese que los escenarios A y B son excluyentes entre sí, pero no A y C. Los escenarios  B y C tampoco son excluyentes entre sí.')
    print('')
    escenarios(CCD,NPADs)
    return(None)

def centroscampus(cen,camp):
    C=[]
    for l in range(len(cen)):
        if cen[l][1]==camp[0]:
            C=C+[cen[l]]
        else:
            C=C
    return(C)
```
